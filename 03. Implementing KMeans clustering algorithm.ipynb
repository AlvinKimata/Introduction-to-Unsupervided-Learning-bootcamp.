```python
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

### Working of the KMeans algorithm.
- The first step is to randomly select k centroids, where k is equal to the number of clusters you choose. Centroids are data points representing the center of a cluster.
- The main element of the algorithm works by a two-step process called expectation-maximization.
-  The expectation step assigns each data point to its nearest centroid.
- Then, the maximization step computes the mean of all the points for each cluster and sets the new centroid.

```python
#A 2D dataset containing 4 distinct blobs.
features, true_labels = make_blobs(n_samples = 300, centers = 4, cluster_std = .6, random_state = 0)

#centers parameter is the number of centers to generate.
# cluster_std is the standard deviation.
```

```python
plt.figure(figsize = (10, 7))
plt.scatter(features[:, 0], features[:, 1], s = 50)
plt.show()
```

```python
#Perform feature scaling.
scaler = StandardScaler()
X = scaler.fit_transform(features)
```

## Build a KMeans algorithm.

```python
##Parameters:
#init; Controls the initialization technique.
#n_clusters; Sets 'k' for the clustering step.
#n_init; Sets the number of initialization steps to perform.
kmeans = KMeans(init = 'random', n_clusters = 4, n_init = 10, max_iter = 300, random_state = 0)
```

```python
#Train the model.
kmeans.fit(X)
```

```python
### Assessing the attributes from training.
kmeans.inertia_
```

```python
#Finding location of centroids.
kmeans.n_iter_
```

```python
y_kmeans = kmeans.predict(X)
```

```python
plt.figure(figsize = (15, 8))
plt.scatter(X[:, 0], X[:, 1], c = y_kmeans, s = 50, cmap = 'viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'black', s = 200, alpha = 0.5)
plt.show()
```
